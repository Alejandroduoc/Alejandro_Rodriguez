```python
print ( "----------------------- RENAPER  Totem Autoservicio-------------------------")
personas=[]


def validar_rut(rut):
    rut = rut.replace("-", "").replace(".", "")
    if len(rut) != 9:
        return False
    if not rut[:-1].isdigit():
        return False

    digito_verificador = rut[-1].upper()
    rut_numeros = list(map(int, rut[:-1]))

    multiplicador = 2
    suma = 0
    for i in range(len(rut_numeros)-1, -1, -1):
        suma += rut_numeros[i] * multiplicador
        multiplicador = (multiplicador + 1) % 8 or 2

    digito_calculado = 11 - (suma % 11)
    digito_calculado = str(digito_calculado) if digito_calculado != 10 else "K"

    return digito_calculado == digito_verificador


def guardar():
  while True:
    rut = input("Ingrese un DNI: ")
    rut= rut.replace("-", "").replace(".", "")
    if validar_rut(rut):
      print("DNI válido")
      nombre = input("Nombre: ")
      if len(nombre) < 8:
       print("El nombre debe tener mínimo 8 caracteres")
       return
      edad = int(input("Edad: "))
      if edad < 0:
       print("La edad debe ser mayor a 0")
       return

      Nacionalidad = input("Nacionalidad: ")
      ciudad_nacimiento = input("Ciudad de nacimiento: ")
      estado_civil=input("Estado civil: ")

      persona = {
        "rut": rut,
        "nombre": nombre,
        "edad": edad,
        "Nacionalidad": Nacionalidad,
        "ciudad_nacimiento": ciudad_nacimiento,
        "estado_civil":estado_civil
        }
      return persona
    else:
     print("DNI no valido")

def buscar(personas):
 while True:
    rut = input("Ingrese un DNI: ")
    rut= rut.replace("-", "").replace(".", "")
    if validar_rut(rut):
      print("DNI válido")
      for persona in personas:
        if persona["rut"] == rut:
          print("Información de la persona:")
          print("DNI: ", persona["rut"])
          print("Nombre: ", persona["nombre"])
          print("Edad: ", persona["edad"])
          print("Nacionalidad: ", persona["Nacionalidad"])
          print("Ciudad de nacimiento: ", persona["ciudad_nacimiento"])
          if persona["Nacionalidad"].lower() == "argentino" or persona["Nacionalidad"].lower() =="argentina":
           print("La persona es de Nacionalidad Argentina")
          else:
           print("La persona no es argentina")
          return persona
    else:
        print("Persona no encontrada")

 else:
     print("DNI no valido ")
def imprimir_certificados(personas):


 print("Certificados de Nacimiento:")
 for persona in personas:


  print("Nombre:", persona["nombre"])
  print("DNI:", persona["rut"])


  print("Nacionalidad:", persona["Nacionalidad"])
  print("Ciudad de nacimiento:", persona["ciudad_nacimiento"])
  print("estado conyugal:", persona["estado_civil"])
  print("-----------------------------")

def certi(personas):
 while True:
    rut = input("Ingrese un DNI: ")
    rut= rut.replace("-", "").replace(".", "")
    if validar_rut(rut):
      print("DNI válido")
      for persona in personas:
              if persona["rut"] == rut:
                 print("Certificado de Nacimiento")
                 print("Nombre: ", persona["nombre"])
                 print("DNI: ", persona["rut"])
                 print("Nacionalidad: ", persona["Nacionalidad"])
                 print("Ciudad de nacimiento: ", persona["ciudad_nacimiento"])
                 print("estado conyugal:", persona["estado_civil"])
                 return persona
              else:
               print("Persona no encontrada")
 else:
  print("DNI no Valido")

def eliminar(personas):
 while True:
    rut = input("Ingrese un DNI: ")
    rut= rut.replace("-", "").replace(".", "")
    if validar_rut(rut):
      print("DNI válido")
      for persona in personas:
        if persona["rut"] == rut:
            personas.remove(persona)
            print("Registro Eliminado ")
            return persona
    else:
        print("Persona no encontrada")

 else:
     print("RUT no valido ")

while True:

        print("----- MENÚ --------------------")
        print("-----1.-Guardar usuario--------")
        print("-----2.-Buscar-----------------")
        print("-----3.-Imprimir certificado por DNI--")
        print("-----4.-Eliminar---------------")
        print("-----5.-Imprimir certificados--")
        print("-----6.-Salir------------------")
        try:
          op=int(input("Ingrese opcion: "))
          while op< 1 or op> 6 :
           print("Ingresar Opcion Valida ")
           op=int(input("Ingrese opcion: "))
          if op==1:
             persona=guardar()
             if persona:
              personas.append(persona)
              print("Registro Guardado")
              print(personas)
          elif op==2:
              buscar(personas)
          elif op==3:
            certi(personas)

          elif op== 4:
            eliminar(personas)
          elif op==6:
           print("Cerrando sistema")
           print("Alejandro Roderiguez")
           print("Version 1001")
           break
          elif op == 5:
           imprimir_certificados(personas)

        except ValueError:
           print("---ingresar solo numeros indicados en el menu-----")






```
