```python
departamentos = []
ganancias_ventas = {
    'F': 0,
    'G': 0,
    'H': 0,
    'I': 0,
    'J': 0
}


def iniciar():
    # Crear los pisos y los departamentos en cada piso
    for piso in range(1, 11):
        pisos = []
        for i in range(5):
            departamento = {
                "DEP": chr(70 + i) + str(piso),  # Convertir números a letras y agregar el número del piso
                'EST': 'D',  # Estado inicial del departamento
                'PRECIOV': 0 , # Precio inicial del departamento
                'PRECIOA': 0 , # Precio inicial del departamento
            }
            if i == 0:  # Tipo F
                departamento['PRECIOV'] = 3800
            elif i == 1:  # Tipo G
                departamento['PRECIOV'] = 3200
            elif i == 2:  # Tipo H
                departamento['PRECIOV'] = 3100
            elif i == 3:  # Tipo I
                departamento['PRECIOV'] = 3000
            elif i == 4:  # Tipo J
                departamento['PRECIOV'] = 3400
            if i == 0:  # Tipo F
                departamento['PRECIOA'] = 25
            elif i == 1:  # Tipo G
                departamento['PRECIOA'] = 18
            elif i == 2:  # Tipo H
                departamento['PRECIOA'] = 16
            elif i == 3:  # Tipo I
                departamento['PRECIOA'] = 23
            elif i == 4:  # Tipo J
                departamento['PRECIOA'] = 28
            pisos.append(departamento)
        departamentos.append(pisos)


def imprimir_departamentos():
    # Imprimir la matriz de departamentos
    for piso in departamentos:
        for departamento in piso:
            estado_departamento = departamento['EST']
            if estado_departamento == 'D':
                print(departamento['DEP'], estado_departamento, "\033[1;32m"+"VENTA:", departamento['PRECIOV'], "UF"+ "\033[0m","\033[1;33m"+"ARRIENDO", departamento['PRECIOA'], "UF"+ "\033[0m", end=" ")
            elif estado_departamento == 'V':
                print( "\033[1;31m" + departamento['DEP'], estado_departamento + "\033[0m","\033[1;33m"+ "ARRIENDO:", departamento['PRECIOA'], "UF" + "\033[0m", end=" ")
            elif estado_departamento == 'A':
                print("\033[1;31m" + departamento['DEP'], estado_departamento + "\033[0m","\033[1;32m"+ "VENTA:", departamento['PRECIOV'], "UF" + "\033[0m", end=" ")
        print()
def validar_rut(rut):
    rut = rut.replace("-", "").replace(".", "")
    if len(rut) != 9:
        return False
    if not rut[:-1].isdigit():
        return False

    digito_verificador = rut[-1].upper()
    rut_numeros = list(map(int, rut[:-1]))

    multiplicador = 2
    suma = 0
    for i in range(len(rut_numeros)-1, -1, -1):
        suma += rut_numeros[i] * multiplicador
        multiplicador = (multiplicador + 1) % 8 or 2

    digito_calculado = 11 - (suma % 11)
    digito_calculado = str(digito_calculado) if digito_calculado != 10 else "K"

    return digito_calculado == digito_verificador
def mostrar_lista_compradores():
    print("\033[1;31m"+"Lista de compradores:" + "\033[0m")
    for piso in departamentos:
        for departamento in piso:
            if departamento.get('COMPRADOR'):
                print("Departamento:", departamento['DEP'])
                print("Nombre:", departamento['COMPRADOR']['NOMBRE'])
                print("RUT:", departamento['COMPRADOR']['RUT'])
                print("--------------------")
def estado():
   nombre_persona = input("Ingrese el nombre de la persona que realiza la compra: ")
   while True:
    rut = input("Ingrese el RUT de la persona que realiza la compra: ")
    rut= rut.replace("-", "").replace(".", "")
    if validar_rut(rut):
      print("RUT válido")
      global ventas_totales  # Declarar la variable ventas_totales como global

      imprimir_departamentos()
      while True:
          nomd = input("Ingrese el departamento a comprar o arrendar: ")
          nombre_departamento = nomd.upper()
          encontrado = False
          for piso in departamentos:
              for departamento in piso:
                  if departamento['DEP'] == nombre_departamento:
                      encontrado = True
                      if departamento['EST'] == 'V' or departamento['EST'] == 'A':
                          print("El departamento no se encuentra disponible. Intente nuevamente.")
                          estado()
                      else:
                          while True:
                              compra = input("Ingrese el nuevo estado del departamento (Venta[V] o arriendo[A]): ")
                              compra = compra.upper()
                              if compra == "V":

                                  nuevo_estado = compra
                                  departamento['EST'] = nuevo_estado
                                  print("El departamento", nombre_departamento, "tiene estado:", "\033[1;31m" + departamento['EST'] + "\033[0m")
                                  ganancias_ventas[departamento['DEP'][0]] += departamento['PRECIOV']

                                  departamento['COMPRADOR'] = {
                                      'NOMBRE': nombre_persona,
                                      'RUT': rut
                                  }
                                  break
                              if compra == 'A':
                                  nuevo_estado = compra
                                  departamento['EST'] = nuevo_estado
                                  print("El departamento", nombre_departamento, "tiene estado:", "\033[1;31m" + departamento['EST'] + "\033[0m")
                                  break
                              else:
                                print("Ingrese V o A")
                      break
          if encontrado:
              break
          else:
           print("El departamento", nombre_departamento, "no se encontró. Intente nuevamente.")

    else:
     print("RUT no valido")
    break
def mostrar_ventas_por_tipo():
    total_cantidad_vendida = 0
    total_ganancia = 0
    print("\033[1;31m"+"Ventas por tipo de departamento:" + "\033[0m")
    for tipo, ganancia in ganancias_ventas.items():
        cantidad_vendida =round( ganancia / departamentos[0][0]['PRECIOV'])  # Calcular la cantidad vendida dividiendo la ganancia entre el precio de un departamento del tipo correspondiente
        print("Tipo", tipo + ":", "Cantidad vendida:", cantidad_vendida, "Total vendido:", ganancia, "UF")
        total_cantidad_vendida += cantidad_vendida
        total_ganancia += ganancia
    print("\033[1;31m"+"Ventas totales:" + "\033[0m")
    print("Cantidad total de departamentos vendidos:", total_cantidad_vendida,"Ganancia total:", total_ganancia, "UF")


def main():
    iniciar()
    while True:
        print("\033[1;31m"+"                     --- Menú ---" + "\033[0m")
        print("1. Comprar departamento o Arrendar departamento")
        print("2. Mostrar departamentos disponibles")
        print("3. Ver listado de compradores")
        print("4. Mostrar ventas de departamento")
        print("5. Salir")

        opcion = input("Ingrese el número de la opción deseada: ")

        if opcion == "1":
            estado()
        elif opcion == "2":
            imprimir_departamentos()
        elif opcion == "3":
            mostrar_lista_compradores()
        elif opcion == "4":
            mostrar_ventas_por_tipo()
        elif opcion == "5":
            break
        else:
            print("Opción inválida. Por favor, ingrese un número válido.")

# Ejecutar la función principal
main()
```
